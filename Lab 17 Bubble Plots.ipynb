### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
# Convert Age to numeric midpoint values
age_mapping = {
    'Under 18 years old': 16, '18-24 years old': 21, '25-34 years old': 29.5,
    '35-44 years old': 39.5, '45-54 years old': 49.5, '55-64 years old': 59.5,
    '65 years or older': 70, 'Prefer not to say': None
}
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Map SOPartFreq to numeric values (approximate participation frequency)
part_freq_map = {
    'Less than once per month': 0.5, 'A few times per month': 2, 
    'A few times per week': 10, 'Daily or almost daily': 30, 
    'Multiple times per day': 50
}
df['PartFreqNum'] = df['SOPartFreq'].map(part_freq_map)

# Filter data for plotting
bubble_df = df[['AgeNumeric', 'PartFreqNum', 'JobSat']].dropna()

# Plot bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(bubble_df['AgeNumeric'], bubble_df['PartFreqNum'], 
            s=bubble_df['JobSat'] * 50, alpha=0.5, c='blue')
plt.title('Age vs. Participation Frequency (Bubble Size: Job Satisfaction)')
plt.xlabel('Age (Approximate Midpoint)')
plt.ylabel('Participation Frequency (Approx. Times per Month)')
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here
# Filter data for plotting
bubble_df = df[['ConvertedCompYearly', 'JobSat', 'AgeNumeric']].dropna()
bubble_df = bubble_df[bubble_df['ConvertedCompYearly'] < 1e7]  # Limit extreme values

# Plot bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(bubble_df['ConvertedCompYearly'], bubble_df['JobSat'], 
            s=bubble_df['AgeNumeric'] * 10, alpha=0.5, c='green')
plt.title('Compensation vs. Job Satisfaction (Bubble Size: Age)')
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Job Satisfaction')
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here
# Split LanguageHaveWorkedWith and calculate counts per language
lang_df = df[['LanguageHaveWorkedWith', 'AgeNumeric']].dropna()
lang_counts = lang_df['LanguageHaveWorkedWith'].str.split(';', expand=True).stack().value_counts()
top_langs = lang_counts.head(10).index  # Top 10 languages

# Calculate mean AgeNumeric for each language
lang_age = {}
for lang in top_langs:
    mask = lang_df['LanguageHaveWorkedWith'].str.contains(lang, na=False)
    lang_age[lang] = lang_df.loc[mask, 'AgeNumeric'].mean()

# Prepare data for bubble plot
langs = list(lang_age.keys())
ages = list(lang_age.values())
sizes = [lang_counts[lang] for lang in langs]  # Size by frequency

# Plot bubble plot
plt.figure(figsize=(12, 6))
plt.scatter(ages, langs, s=[s * 10 for s in sizes], alpha=0.5, c='purple')
plt.title('Technology Preferences by Age (Bubble Size: Frequency)')
plt.xlabel('Average Age (Approximate Midpoint)')
plt.ylabel('Programming Language')
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here
# Split DatabaseWantToWorkWith and calculate mean JobSat
db_df = df[['DatabaseWantToWorkWith', 'JobSat']].dropna()
db_counts = db_df['DatabaseWantToWorkWith'].str.split(';', expand=True).stack().value_counts()
top_dbs = db_counts.head(10).index  # Top 10 databases

# Calculate mean JobSat for each database
db_jobsat = {}
for db in top_dbs:
    mask = db_df['DatabaseWantToWorkWith'].str.contains(db, na=False)
    db_jobsat[db] = db_df.loc[mask, 'JobSat'].mean()

# Prepare data for bubble plot
dbs = list(db_jobsat.keys())
jobsat_values = list(db_jobsat.values())
sizes = [db_counts[db] for db in dbs]  # Size by number of respondents

# Plot bubble plot
plt.figure(figsize=(12, 6))
plt.scatter(jobsat_values, dbs, s=[s * 10 for s in sizes], alpha=0.5, c='orange')
plt.title('Preferred Databases vs. Job Satisfaction (Bubble Size: Respondents)')
plt.xlabel('Average Job Satisfaction')
plt.ylabel('Database')
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here
# Split DevType and calculate mean ConvertedCompYearly
dev_df = df[['DevType', 'ConvertedCompYearly', 'JobSat']].dropna()
dev_df = dev_df[dev_df['ConvertedCompYearly'] < 1e7]
dev_counts = dev_df['DevType'].str.split(';', expand=True).stack().value_counts()
top_devs = dev_counts.head(5).index  # Top 5 developer types

# Calculate mean compensation and JobSat for each dev type
dev_comp = {}
dev_jobsat = {}
for dev in top_devs:
    mask = dev_df['DevType'].str.contains(dev, na=False)
    dev_comp[dev] = dev_df.loc[mask, 'ConvertedCompYearly'].mean()
    dev_jobsat[dev] = dev_df.loc[mask, 'JobSat'].mean()

# Prepare data for bubble plot
devs = list(dev_comp.keys())
comp_values = list(dev_comp.values())
sizes = [dev_jobsat[dev] * 50 for dev in devs]  # Size by JobSat

# Plot bubble plot
plt.figure(figsize=(12, 6))
plt.scatter(comp_values, devs, s=sizes, alpha=0.5, c='red')
plt.title('Compensation Across Developer Roles (Bubble Size: Job Satisfaction)')
plt.xlabel('Average Yearly Compensation (USD)')
plt.ylabel('Developer Role')
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here
# Split NEWCollabToolsHaveWorkedWith and calculate mean AgeNumeric
collab_df = df[['NEWCollabToolsHaveWorkedWith', 'AgeNumeric']].dropna()
collab_counts = collab_df['NEWCollabToolsHaveWorkedWith'].str.split(';', expand=True).stack().value_counts()
top_tools = collab_counts.head(10).index  # Top 10 tools

# Calculate mean AgeNumeric for each tool
tool_age = {}
for tool in top_tools:
    mask = collab_df['NEWCollabToolsHaveWorkedWith'].str.contains(tool, na=False)
    tool_age[tool] = collab_df.loc[mask, 'AgeNumeric'].mean()

# Prepare data for bubble plot
tools = list(tool_age.keys())
ages = list(tool_age.values())
sizes = [collab_counts[tool] for tool in tools]  # Size by frequency

# Plot bubble plot
plt.figure(figsize=(12, 6))
plt.scatter(ages, tools, s=[s * 10 for s in sizes], alpha=0.5, c='cyan')
plt.title('Collaboration Tools by Age (Bubble Size: Frequency)')
plt.xlabel('Average Age (Approximate Midpoint)')
plt.ylabel('Collaboration Tool')
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here
# Split WebframeWantToWorkWith and calculate mean JobSat
web_df = df[['WebframeWantToWorkWith', 'JobSat']].dropna()
web_counts = web_df['WebframeWantToWorkWith'].str.split(';', expand=True).stack().value_counts()
top_webs = web_counts.head(10).index  # Top 10 frameworks

# Calculate mean JobSat for each framework
web_jobsat = {}
for web in top_webs:
    mask = web_df['WebframeWantToWorkWith'].str.contains(web, na=False)
    web_jobsat[web] = web_df.loc[mask, 'JobSat'].mean()

# Prepare data for bubble plot
webs = list(web_jobsat.keys())
jobsat_values = list(web_jobsat.values())
sizes = [web_counts[web] for web in webs]  # Size by number of respondents

# Plot bubble plot
plt.figure(figsize=(12, 6))
plt.scatter(jobsat_values, webs, s=[s * 10 for s in sizes], alpha=0.5, c='magenta')
plt.title('Preferred Web Frameworks vs. Job Satisfaction (Bubble Size: Respondents)')
plt.xlabel('Average Job Satisfaction')
plt.ylabel('Web Framework')
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
##Write your code here
# Split LanguageAdmired and calculate counts per country
admire_df = df[['LanguageAdmired', 'Country']].dropna()
admire_counts = admire_df['LanguageAdmired'].str.split(';', expand=True).stack().value_counts()
top_langs = admire_counts.head(5).index  # Top 5 admired languages
top_countries = admire_df['Country'].value_counts().head(5).index  # Top 5 countries

# Filter for top countries and languages
admire_df = admire_df[admire_df['Country'].isin(top_countries)]

# Calculate counts for each language-country pair
lang_country_counts = {}
for country in top_countries:
    for lang in top_langs:
        mask = (admire_df['Country'] == country) & (admire_df['LanguageAdmired'].str.contains(lang, na=False))
        lang_country_counts[(country, lang)] = admire_df[mask].shape[0]

# Prepare data for bubble plot
countries = [pair[0] for pair in lang_country_counts.keys()]
langs = [pair[1] for pair in lang_country_counts.keys()]
sizes = list(lang_country_counts.values())

# Plot bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(countries, langs, s=[s * 10 for s in sizes], alpha=0.5, c='teal')
plt.title('Admired Technologies Across Countries (Bubble Size: Frequency)')
plt.xlabel('Country')
plt.ylabel('Admired Language')
plt.xticks(rotation=45, ha='right')
plt.show()
```
