#### Step 1: Import the necessary module.


### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

#### 2. Explore the Dataset


<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
# Write your code here
# Display dataset info including data types and missing values
print("Dataset Info:")
print(df.info())

# Display count of missing values per column
print("\nMissing Values per Column:")
print(df.isnull().sum())
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
# Write your code here
# Generate descriptive statistics for numerical columns
print("Basic Statistics for Numerical Columns:")
print(df.describe())
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
# Check unique values in the Country column to identify inconsistencies
print("Unique values in Country column:")
print(df['Country'].unique())

# Count occurrences of each unique value
print("\nValue counts for Country:")
print(df['Country'].value_counts())
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
# Example: Standardize Country names (assuming some inconsistencies like 'USA' vs 'United States')
country_mapping = {
    'United States of America': 'USA',
    'United States': 'USA',
    # Add more mappings as needed based on exploration
}
df['Country'] = df['Country'].replace(country_mapping)

# Standardize EdLevel (example mapping)
edlevel_mapping = {
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': 'Bachelor',
    'Master’s degree (M.A., M.S., M.Eng., etc.)': 'Master',
    'Something else': 'Other',
    # Add more mappings as needed
}
df['EdLevel'] = df['EdLevel'].replace(edlevel_mapping)

# Verify changes
print("Standardized Country values:\n", df['Country'].value_counts())
print("\nStandardized EdLevel values:\n", df['EdLevel'].value_counts())
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
# Apply one-hot encoding to Employment column
df_encoded = pd.get_dummies(df, columns=['Employment'], prefix='Employment')

# Display the first few rows of the encoded dataframe
print("Dataframe with one-hot encoded Employment:\n", df_encoded.head())
df = df_encoded  # Update the original dataframe
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
# Calculate missing values per column and sort
missing_values = df.isnull().sum().sort_values(ascending=False)
print("Columns with the highest number of missing values:\n", missing_values.head(10))
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
# Impute missing values in ConvertedCompYearly with the median
median_comp = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(median_comp)

# Verify imputation
print("Missing values in ConvertedCompYearly after imputation:", df['ConvertedCompYearly'].isnull().sum())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
# Impute missing values in RemoteWork with the most frequent value
most_frequent_remote = df['RemoteWork'].mode()[0]
df['RemoteWork'] = df['RemoteWork'].fillna(most_frequent_remote)

# Verify imputation
print("Missing values in RemoteWork after imputation:", df['RemoteWork'].isnull().sum())
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
# Apply Min-Max Scaling to ConvertedCompYearly
min_comp = df['ConvertedCompYearly'].min()
max_comp = df['ConvertedCompYearly'].max()
df['ConvertedCompYearly_MinMax'] = (df['ConvertedCompYearly'] - min_comp) / (max_comp - min_comp)

# Verify the normalization
print("Min-Max Normalized ConvertedCompYearly (first 5 rows):\n", df['ConvertedCompYearly_MinMax'].head())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
import numpy as np

# Apply log transformation (add 1 to avoid log(0))
df['ConvertedCompYearly_Log'] = np.log1p(df['ConvertedCompYearly'])

# Verify the transformation
print("Log-transformed ConvertedCompYearly (first 5 rows):\n", df['ConvertedCompYearly_Log'].head())
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
# Convert YearsCodePro to numeric, handling non-numeric values
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Define ExperienceLevel based on YearsCodePro
def categorize_experience(years):
    if pd.isna(years):
        return 'Unknown'
    elif years < 2:
        return 'Junior'
    elif years < 5:
        return 'Mid-level'
    else:
        return 'Senior'

df['ExperienceLevel'] = df['YearsCodePro'].apply(categorize_experience)

# Verify the new column
print("ExperienceLevel distribution:\n", df['ExperienceLevel'].value_counts())
```
