##### Import Necessary Modules:


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Tasks


<h2>1. Load the Dataset</h2>
<p>
We use the <code>pandas.read_csv()</code> function for reading CSV files. However, in this version of the lab, which operates on JupyterLite, the dataset needs to be downloaded to the interface using the provided code below.
</p>


The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
# Display basic information about the dataset
print("Dataset Information:")
print(df.info())

# Display summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Display the column names
print("\nColumn Names:")
print(df.columns.tolist())

# Display the shape of the dataset (rows, columns)
print("\nDataset Shape:")
print(df.shape)
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
# Check for missing values in each column
print("Missing Values Count for Each Column:")
print(df.isnull().sum())

# Calculate the percentage of missing values
print("\nPercentage of Missing Values:")
print(df.isnull().sum() / len(df) * 100)

# Total missing values in the dataset
print("\nTotal Missing Values in the Dataset:")
print(df.isnull().sum().sum())
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
# Create a heatmap to visualize missing values
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Values Heatmap')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
# Count missing values in the Employment column
missing_employment = df['Employment'].isnull().sum()
print(f"Number of missing values in Employment column: {missing_employment}")
print(f"Percentage of missing values: {missing_employment / len(df) * 100:.2f}%")

# Display rows with missing Employment values
print("\nSample of rows with missing Employment values:")
print(df[df['Employment'].isnull()].head())
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
# Find the most frequent value in the Employment column
most_frequent_employment = df['Employment'].mode()[0]
print(f"Most frequent value in Employment column: {most_frequent_employment}")

# Display the value counts for the Employment column
print("\nValue counts for Employment:")
print(df['Employment'].value_counts())
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
# Create a copy of the dataframe to preserve the original
df_imputed = df.copy()

# Impute missing values with the most frequent value
most_frequent_employment = df['Employment'].mode()[0]
df_imputed['Employment'].fillna(most_frequent_employment, inplace=True)

# Verify imputation was successful
print("Missing values in Employment column after imputation:")
print(df_imputed['Employment'].isnull().sum())

# Compare before and after imputation
print("\nBefore imputation:")
print(df['Employment'].value_counts())
print("\nAfter imputation:")
print(df_imputed['Employment'].value_counts())
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
# Visualize the distribution of Employment before and after imputation
plt.figure(figsize=(14, 6))

# Before imputation
plt.subplot(1, 2, 1)
df['Employment'].value_counts().plot(kind='bar')
plt.title('Employment Distribution (Before Imputation)')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Count')

# After imputation
plt.subplot(1, 2, 2)
df_imputed['Employment'].value_counts().plot(kind='bar')
plt.title('Employment Distribution (After Imputation)')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

# Alternative: Create a pie chart to visualize the distribution after imputation
plt.figure(figsize=(10, 8))
df_imputed['Employment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Employment Distribution After Imputation')
plt.ylabel('')
plt.show()
```
